# 일변수 벡터함수 3차원으로 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.set_zlabel('$z$', fontsize=20)

t = np.linspace(0, 2, 101)
x = np.sin(6*t)
y = 1/4 * t
z = t**2 / 2

ax.plot3D(x,y,z, c='k')
ax.plot([x[0]], [y[0]], [z[0]], 'o', markersize=10, color='k', label="t = {:.2f}".format(t[0]))
ax.plot([x[50]], [y[50]], [z[50]], 'v', markersize=10, color='k', label="t = {:.2f}".format(t[50]))
ax.plot([x[-1]], [y[-1]], [z[-1]], '*', markersize=10, color='k', label="t = {:.2f}".format(t[-1]))

ax.legend(fontsize=15, loc="upper left")

plt.show()
```

# 벡터변수 스칼라함수 그리기
- 입력은 벡터 1개(그 안에 요소가 여러개)
- 출력은 스칼라 1개

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel(r'$x_1$', fontsize=20)
ax.set_ylabel(r'$y_1$', fontsize=20)
ax.set_zlabel(r'$z_1$', fontsize=20)

x1 = np.linspace(-2, 2, 50)
x2 = np.linspace(-1, 3, 50)
X1, X2 = np.meshgrid(x1, x2)
Z = 50*(X2 - X1**2) + (2-X1)**2

ax.scatter3D(X1, X2, Z, marker='.', color='gray')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel(r'$x_1$', fontsize=20)
ax.set_ylabel(r'$y_1$', fontsize=20)
ax.set_zlabel(r'$z_1$', fontsize=20)

x1 = np.linspace(-2, 2, 50)
x2 = np.linspace(-1, 3, 50)
X1, X2 = np.meshgrid(x1, x2)
Z = 50*(X2 - X1**2) + (2-X1)**2

ax.plot_surface(X1, X2, Z, cmap=plt.cm.binary, edgecolor='k') #곡면으로 그려보기 위해 plot 함수 변경

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(7,7))
ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel(r'$x$', fontsize=20)
ax.set_ylabel(r'$y$', fontsize=20)
ax.set_zlabel(r'$z$', fontsize=20)

u = np.linspace(0, 1, 30)
v = np.linspace(0, 1, 30)
U, V = np.meshgrid(u, v)
X = U
Y = V
Z = (1+U**2) + (V/(1+V**2))

ax.scatter3D(X,Y,Z, marker='.', color='gray')

plt.show()
```
