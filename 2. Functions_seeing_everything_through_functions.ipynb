#### arrowed_spines 함수 세팅

```python
def arrowed_spines(fig, ax, remove_ticks=False):
    """
    좌표축 화살표를 그리기 위한 함수
    https://stackoverflow.com/questions/33737736/matplotlib-axis-arrow-tip
    """
    xmin, xmax = ax.get_xlim() 
    ymin, ymax = ax.get_ylim()

    # removing the default axis on all sides:
    for side in ['bottom','right','top','left']:
        ax.spines[side].set_visible(False)

    if remove_ticks == True: 
        # removing the axis ticks
        plt.xticks([]) # labels 
        plt.yticks([])
        ax.xaxis.set_ticks_position('none') # tick markers
        ax.yaxis.set_ticks_position('none')

    # get width and height of axes object to compute 
    # matching arrowhead length and width
    dps = fig.dpi_scale_trans.inverted()
    bbox = ax.get_window_extent().transformed(dps)
    width, height = bbox.width, bbox.height

    # manual arrowhead width and length
    hw = 1./50.*(ymax-ymin) 
    hl = 1./25.*(xmax-xmin)
    lw = 1. # axis line width
    ohg = 0.4 # arrow overhang

    # compute matching arrowhead length and width
    yhw = hw/(ymax-ymin)*(xmax-xmin)* height/width 
    yhl = hl/(xmax-xmin)*(ymax-ymin)* width/height

    # draw x and y axis
    ax.arrow(xmin, 0, xmax-xmin, 0., fc='k', ec='k', lw = lw, 
             head_width=hw, head_length=hl, #overhang = ohg, 
             length_includes_head= True, clip_on = False) 

    ax.arrow(0, ymin, 0., ymax-ymin, fc='k', ec='k', lw = lw, 
             head_width=yhw, head_length=yhl, #overhang = ohg, 
             length_includes_head= True, clip_on = False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.style.use('bmh')
mpl.style.use('seaborn-whitegrid')

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(1,1,1)
ax.xaxis.set_tick_params(labelsize=18)
ax.yaxis.set_tick_params(labelsize=18)
plt.xlabel('$x$', fontsize=25)
plt.ylabel('$y$', fontsize=25)

x = np.linspace(-3, 2, 10)
y = 2*x + 4
ax.plot(x,y,'k')

arrowed_spines(fig,ax)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.style.use('bmh')
mpl.style.use('seaborn-whitegrid')

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1,2,1)
ax.xaxis.set_tick_params(labelsize=18)
ax.yaxis.set_tick_params(labelsize=18)
ax.set_xlabel('$x$', fontsize=25)
ax.set_ylabel('$y$', fontsize=25)

x = np.linspace(-2, 5, 100)
y = 2**x
ax.plot(x,y,'k')
ax.plot(x, 3**x, '--')
ax.plot(x, 4**x, '.')

ax2 = fig.add_subplot(1,2,2)
ax2.xaxis.set_tick_params(labelsize=18)
ax2.yaxis.set_tick_params(labelsize=18)
ax2.set_xlabel('$x$', fontsize=25)
ax2.set_ylabel('$y$', fontsize=25)

x = np.linspace(-2, 5, 1000)
y = (1/2)**x
ax2.plot(x,y,'k')


arrowed_spines(fig,ax)
arrowed_spines(fig,ax2)



plt.show()
```

## 로그함수 표현하기

```python
import numpy as np

def log(x, base=np.e):
    return np.log(x) / np.log(base)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.style.use('bmh')
mpl.style.use('seaborn-whitegrid')

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1,2,1)
ax.xaxis.set_tick_params(labelsize=18)
ax.yaxis.set_tick_params(labelsize=18)
ax.set_xlabel('$x$', fontsize=25)
ax.set_ylabel('$y$', fontsize=25)

x = np.linspace(0.00001, 5, 1000)
y = log(x, 10)
ax.plot(x,y,'k', label='log(10,x)')
ax.plot(x, np.log(x), '--', label='log(e,x)')
ax.legend(fontsize=20)

ax2 = fig.add_subplot(1,2,2)
ax2.xaxis.set_tick_params(labelsize=18)
ax2.yaxis.set_tick_params(labelsize=18)
ax2.set_xlabel('$x$', fontsize=25)
ax2.set_ylabel('$y$', fontsize=25)

x = np.linspace(0.00001, 5, 1000)
y = log(x, (1/10))
ax2.plot(x,y,'k', label='log(1/10, x)')
ax2.plot(x, log(x, (1/np.e)), '--', label='log(1/e, x)')
ax2.legend(fontsize=20)

arrowed_spines(fig,ax)
arrowed_spines(fig,ax2)



plt.show()
```
